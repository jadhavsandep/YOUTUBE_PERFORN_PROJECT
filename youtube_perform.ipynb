```python
data.groupby('Year')['Estimated Revenue (USD)'].sum().plot(
    kind='line', marker='o', figsize=(10, 6), title='Yearly Revenue Trends'
)
plt.xlabel('Year')
plt.ylabel('Total Revenue (USD)')
plt.grid(True)
plt.show()
```

```python
data.groupby('Year')['Views'].sum().plot.pie(
    autopct='%1.1f%%', figsize=(8, 8), title='Distribution of Views by Year', startangle=90
)
plt.ylabel('')  # Remove default ylabel
plt.show()
```

```python
data.groupby('Year')['Subscribers'].sum().plot(
    kind='line', marker='o', color='green', figsize=(10, 6), title='Subscriber Growth Trends by Year'
)
plt.xlabel('Year')
plt.ylabel('Total Subscribers')
plt.grid(True)
plt.show()
```

```python
data.groupby('Year')['Views'].sum().plot(
    kind='bar', figsize=(10, 6), title='Total Views by Year'
)
plt.xlabel('Year')
plt.ylabel('Total Views')
plt.show()
```

```python
data.groupby('Year')['Watch Time (hours)'].sum().plot(
    kind='line', marker='o', figsize=(10, 6), title='Watch Time Trends by Year'
)
plt.xlabel('Year')
plt.ylabel('Watch Time (hours)')
plt.grid(True)
plt.show()
```

```python
data.groupby('Year')['Estimated Revenue (USD)'].sum().plot.pie(
    autopct='%1.1f%%', figsize=(8, 8), title='Revenue Contribution by Year', startangle=90
)
plt.ylabel('')
plt.show()
```

```python
data['Video Duration'].plot(
    kind='hist', bins=30, figsize=(10, 6), title='Distribution of Video Durations'
)
plt.xlabel('Video Duration (seconds)')
plt.ylabel('Frequency')
plt.show()
```

```python
data.plot.scatter(
    x='Views', y='Estimated Revenue (USD)', figsize=(10, 6), title='Views vs Revenue'
)
plt.xlabel('Views')
plt.ylabel('Revenue (USD)')
plt.show()
```

```python
data[['Revenue per 1000 Views (USD)']].plot.box(
    figsize=(8, 6), title='Revenue per 1000 Views Distribution'
)
plt.ylabel('Revenue (USD)')
plt.show()
```

```python
data.groupby('Year')['Subscribers'].sum().plot(
    kind='bar', figsize=(10, 6), title='Total Subscribers by Year', color='orange'
)
plt.xlabel('Year')
plt.ylabel('Subscribers')
plt.show()
```

```python
data.plot.scatter(
    x='Year', y='Watch Time (hours)', figsize=(10, 6), title='Scatter Plot of Watch Time by Year'
)
plt.xlabel('Year')
plt.ylabel('Watch Time (hours)')
plt.show()
```

```python
data.plot.scatter(
    x='Views', y='Estimated Revenue (USD)', figsize=(10, 6), title='Relationship between Views and Revenue'
)
plt.xlabel('Views')
plt.ylabel('Revenue (USD)')
plt.grid(True)
plt.show()
```

```python
data.plot.scatter(
    x='Watch Time (hours)', y='Estimated Revenue (USD)', figsize=(10, 6), title='Revenue Trends against Watch Time'
)
plt.xlabel('Watch Time (hours)')
plt.ylabel('Estimated Revenue (USD)')
plt.grid(True)
plt.show()
```

```python
monthly_revenue = data.groupby('Month')['Estimated Revenue (USD)'].sum()
colors = ['green', 'red', 'yellow'] * (len(monthly_revenue) // 3 + 1)

monthly_revenue.plot(
    kind='bar', color=colors[:len(monthly_revenue)], figsize=(10, 6), title='Revenue Trends Across Different Months'
)
plt.xlabel('Month')
plt.ylabel('Total Revenue (USD)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
video_revenue = data.groupby('ID')['Estimated Revenue (USD)'].sum().sort_values(ascending=False)
colors = ['yellow', 'red'] * 5

video_revenue.head(10).plot(
    kind='bar', color=colors, figsize=(10, 6), title='Top 10 Videos by Revenue'
)
plt.xlabel('Video ID')
plt.ylabel('Total Revenue (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Filter data for the last year in the dataset
last_year = data['Year'].max()
last_year_revenue = data[data['Year'] == last_year].groupby('Month')['Estimated Revenue (USD)'].sum()

# Plot revenue trends for the last year
last_year_revenue.plot(
    kind='line', marker='o', color='blue', figsize=(10, 6), title=f'Revenue Trends for {last_year}'
)
plt.xlabel('Month')
plt.ylabel('Total Revenue (USD)')
plt.grid(True)
plt.show()
```

```python
most_viewed_videos = data.groupby('ID')['Views'].sum().nlargest(10).index
views_over_time = data[data['ID'].isin(most_viewed_videos)].groupby(['Year', 'ID'])['Views'].sum().unstack()

views_over_time.plot(
    kind='line', marker='o', figsize=(12, 8), title='Most Viewed Videos Over Time'
)
plt.xlabel('Year')
plt.ylabel('Views')
plt.legend(title='Video ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Summary of Key Insights
# 1. **Top-performing years**: 2017 had the highest revenue, driven by large viewership.
# 2. **Revenue vs. Views**: Strong correlation; however, some videos generated high revenue with fewer views.
# 3. **Subscriber Trends**: Subscriber growth peaked in specific years and aligned with revenue trends.
# 4. **Monthly Revenue**: Certain months, like December, consistently contributed more to revenue.
```
